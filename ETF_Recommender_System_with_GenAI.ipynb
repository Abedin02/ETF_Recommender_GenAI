#### 6. MongoDB Document Design Principles

Different sources provide different fields. Finviz gives you price, volume, and change. StockAnalysis gives you Asset Class and AUM. Regardless of source, your documents must follow these rules:

**Key normalization requirements:**
- All keys must be **single-word with no spaces or special characters** â€” pick either `snake_case` or `camelCase` and be consistent across your entire project
  - âŒ `'Asset Class'`, `'Fund Name'`, `'%Change'`
  - âœ… `'asset_class'`, `'fund_name'`, `'pct_change'` (snake_case)
  - âœ… `'assetClass'`, `'fundName'`, `'pctChange'` (camelCase)
- MongoDB cannot index or query keys with dots (`.`) or dollar signs (`$`) â€” clean these out
- Numeric fields must be stored as **numbers, not strings** â€” `2150000000.0` not `'$2.15B'`
- Every document must include `etf_ticker` and `query_date`

**The two most important fields are `query_date` and `aum`**. In finance, data without a date is useless â€” you need to know *when* something was true, not just *what* was true. The `query_date` lets you track how ETF AUM evolves over time, which is critical for any time-series analysis. The `aum` field is what you'll use to filter for large Equity ETFs. Make sure to parse strings like `'$2.15B'` and `'$517.3M'` into numeric values so MongoDB can perform range queries and sorts.

**In your presentation, discuss:**
- **Without index**: MongoDB performs a **COLLSCAN** (collection scan) â€” it reads every single document to check if `aum >= 2e9`. If you have 3,000 ETFs Ã— 30 days = 90,000 documents, it examines all 90,000.
- **With index**: MongoDB performs an **IXSCAN** (index scan) â€” it jumps directly to the relevant range. It might examine only 200 documents.
- **The tradeoff**: Indexes speed up reads but slow down writes (every `insert` or `update` must also update the index) and consume additional storage. For a data warehouse where you load once and query many times, this is almost always worth it.
- Show the `winningPlan` from both explain outputs to illustrate COLLSCAN vs IXSCAN.

### Bulk Operations vs Individual Inserts
```python
# âŒ Slow - individual inserts
for doc in documents:
    collection.insert_one(doc)

# âœ… Fast - bulk insert
collection.insert_many(documents)

# âœ… Best - upsert to handle duplicates with date
from datetime import datetime

today_str = datetime.now().strftime('%Y%m%d')

for doc in documents:
    doc['query_date'] = today_str
    collection.update_one(
        {'etf_ticker': doc['etf_ticker'], 'query_date': today_str},
        {'$set': doc},
        upsert=True
    )
```

### Benefits of this approach:
1. **Prevents duplicate daily loads**: The unique index on (query_date, etf_ticker) ensures only one entry per ETF per day
2. **Historical tracking**: Can maintain price history over multiple days
3. **Efficient resumption**: Easy to check which ETFs already have today's data
4. **Clear data freshness**: String date format makes it obvious when data was collected

### Self-Healing Pipeline (REQUIRED)

Your pipeline must handle LLM rate limits gracefully. The pattern: load all holdings first, then cure missing summaries in a separate pass. If you hit a rate limit, stop and resume later â€” the pipeline picks up where it left off.

```python
import time

def generate_missing_summaries(etf_holdings_col, model, today_str, max_calls=50):
    """
    Self-healing summary generator.
    Finds ETFs missing strategy_summary and fills them in,
    respecting rate limits. Run repeatedly until all are filled.
    """

    # Find documents that have holdings but no summary yet
    missing = etf_holdings_col.find({
        'query_date': today_str,
        '$or': [
            {'strategy_summary': None},
            {'strategy_summary': {'$exists': False}}
        ]
    })

    filled = 0
    for doc in missing:
        if filled >= max_calls:
            print(f"Reached {max_calls} calls. Run again later to continue.")
            break

        etf_ticker = doc['etf_ticker']
        holdings = doc['holdings']

        try:
            prompt = build_summary_prompt(etf_ticker, holdings)
            response = model.generate_content(prompt)
            summary = response.text.strip()

            # Update the document with the summary
            etf_holdings_col.update_one(
                {'etf_ticker': etf_ticker, 'query_date': today_str},
                {'$set': {'strategy_summary': summary}}
            )

            filled += 1
            print(f"[{filled}] {etf_ticker}: {summary[:80]}...")

            # Respect rate limits: ~5-10 requests per minute for free tier
            time.sleep(8)

        except Exception as e:
            if '429' in str(e) or 'quota' in str(e).lower():
                print(f"Rate limited at {etf_ticker}. Stopping. Run again later.")
                break
            else:
                print(f"Error on {etf_ticker}: {e}. Skipping.")
                continue

    # Report progress
    total = etf_holdings_col.count_documents({'query_date': today_str})
    done = etf_holdings_col.count_documents({
        'query_date': today_str,
        'strategy_summary': {'$ne': None, '$exists': True}
    })
    print(f"\nProgress: {done}/{total} ETFs have summaries.")
```

````python
import requests
import random
from requests.exceptions import RequestException, Timeout
from bs4 import BeautifulSoup
import pandas as pd
import time
from random_user_agent.user_agent import UserAgent
from random_user_agent.params import SoftwareName, OperatingSystem

software_names = [SoftwareName.CHROME.value]
operating_systems = [OperatingSystem.WINDOWS.value, OperatingSystem.LINUX.value]

user_agent_rotator = UserAgent(software_names=software_names, operating_systems=operating_systems, limit=100)

# Get list of user agents.
user_agents = user_agent_rotator.get_user_agents()

# Get Random User Agent String.
user_agent = user_agent_rotator.get_random_user_agent()

user_agent_rotator.get_random_user_agent()

session = requests.Session()
my_cache = {}

def getETFMaster(page, retry=5):
    if page in my_cache:
        return my_cache[page]

    rpage = page * 20 + 1
    url = f'https://finviz.com/screener.ashx?v=191&f=ind_exchangetradedfund&r={rpage}'

    try:
        # NETWORK REQUEST
        response = session.get(
            url,
            headers={'User-Agent': user_agent_rotator.get_random_user_agent()},
            timeout=10  # 10-second timeout
        )

        # Handle common HTTP errors
        if response.status_code == 403:
            raise RequestException("403 Forbidden")
        elif response.status_code == 429:
            raise RequestException("429 Too Many Requests")
        elif response.status_code != 200:
            raise RequestException(f"HTTP {response.status_code}")

        # random delay
        time.sleep(random.uniform(1.5, 3.0))

        # PARSING

        soup = BeautifulSoup(response.text, 'html.parser')

        # Try to find table (robust to minor HTML changes)
        table = soup.find('table', class_='screener_table')
        if table is None:
            # Sometimes Finviz adds extra tables or changes class names
            tables = soup.find_all('table')
            if not tables:
                print(f"Page {page}: No tables found.")
                return None
            # Try to pick the largest table (most rows)
            table = max(tables, key=lambda t: len(t.find_all('tr')))
            print(f"Page {page}: Using largest table (fallback).")

        # Extract headers
        header_row = table.find('tr')
        if header_row is None:
            print(f"Page {page}: No header row found.")
            return None

        header = [cell.text.strip() for cell in header_row.find_all('th')]
        if not header:
            # Some tables may use <td> for headers
            header = [cell.text.strip() for cell in header_row.find_all('td')]
        if not header:
            print(f"Page {page}: No headers found. Skipping page.")
            return None

        # Extract data rows
        rows = table.find_all('tr')[1:]
        data = []
        for row in rows:
            cells = [cell.text.strip() for cell in row.find_all('td')]
            if cells:  # skip empty rows
                data.append(cells)

        if not data:
            print(f"Page {page}: Table has no data. Skipping page.")
            return None

        # Create DataFrame
        df = pd.DataFrame(data, columns=header)

        # Cache and return
        my_cache[page] = df
        return df

    except (RequestException, Timeout) as e:
        if retry > 0:
            wait_time = random.uniform(2, 5)
            print(f"Page {page}: Network error ({e}). Retrying in {wait_time:.1f} sec...")
            time.sleep(wait_time)
            return getETFMaster(page, retry - 1)
        else:
            print(f"Page {page}: Failed after multiple retries. Skipping...")
            return None
    except Exception as e:
        # Catch all parsing errors
        print(f"Page {page}: Parsing error ({e}). Skipping page.")
        return None
````

````python
import numpy as np
import numpy as np

def convert_value(value):
    """
    Convert value with suffixes to float:
    - B (billions): multiply by 1,000,000,000
    - M (millions): multiply by 1,000,000
    - % (percent): multiply by 0.01
    """
    # Handle NaN, None, empty strings
    if pd.isna(value) or value == '' or value == '-':
        return np.nan

    # Convert to string and clean
    value = str(value).strip().upper()
    value = value.replace('$', '').replace(',', '').replace(' ', '')

    try:
        # Handle billions
        if 'B' in value:
            number = float(value.replace('B', ''))
            return number * 1_000_000_000

        # Handle millions
        elif 'M' in value:
            number = float(value.replace('M', ''))
            return number * 1_000_000

        # Handle percentages
        elif '%' in value:
            number = float(value.replace('%', ''))
            return number * 0.01

        # No suffix, just convert to float
        else:
            return float(value)

    except (ValueError, AttributeError):
        return np.nan

# Apply conversion to all columns except 'Ticker'
non_numeric_cols = ['etf_ticker']
for col in all_etf.columns:
    if col not in non_numeric_cols:
        all_etf[col] = all_etf[col].apply(convert_value)
````

````python
def build_summary_prompt(etf_ticker, holdings):
    top_holdings = holdings[:15]
    holdings_str = "\n".join(
        f"  {h['stock_ticker']}: {h['weight']:.1f}%"
        for h in top_holdings
    )

    prompt = f"""You are a financial analyst writing for investment advisors.
Given the top holdings of ETF '{etf_ticker}', write a 2-3 sentence summary
of the fund's likely investment strategy. Focus on:
- What sectors or themes the holdings represent
- Whether the fund appears concentrated or diversified
- The general market cap tilt (large-cap, mid-cap, small-cap)

Top holdings:
{holdings_str}

Write only the summary, no preamble."""

    return prompt
````

````python
def generate_missing_summaries(etf_holdings_col, model, today_str, max_calls=50):
    missing = etf_holdings_col.find({
        'query_date': today_str,
        '$or': [
            {'strategy_summary': None},
            {'strategy_summary': {'$exists': False}}
        ]
    })

    filled = 0
    for doc in missing:
        if filled >= max_calls:
            print(f"Reached {max_calls} calls. Run again later to continue.")
            break

        etf_ticker = doc['etf_ticker']
        holdings = doc['holdings']

        try:
            prompt = build_summary_prompt(etf_ticker, holdings)
            response = model.generate_content(prompt)
            summary = response.text.strip()

            etf_holdings_col.update_one(
                {'etf_ticker': etf_ticker, 'query_date': today_str},
                {'$set': {'strategy_summary': summary}}
            )

            filled += 1
            print(f"[{filled}] {etf_ticker}: {summary[:80]}...")

            time.sleep(8)

        except Exception as e:
            if '429' in str(e) or 'quota' in str(e).lower():
                print(f"Rate limited at {etf_ticker}. Stopping. Run again later.")
                break
            else:
                print(f"Error on {etf_ticker}: {e}. Skipping.")
                continue

    total = etf_holdings_col.count_documents({'query_date': today_str})
    done = etf_holdings_col.count_documents({
        'query_date': today_str,
        'strategy_summary': {'$ne': None, '$exists': True}
    })
    print(f"\nProgress: {done}/{total} ETFs have summaries.")
````

````python

import re
import math
import numpy as np
import pandas as pd
from collections import Counter, defaultdict
from IPython.display import display
from sentence_transformers import SentenceTransformer


if "etf_holdings_col" not in globals():
    print("etf_holdings_col is not defined.")
    print("Run the MongoDB setup cell first, then rerun this cell.")
    raise SystemExit

if "etf_master_col" not in globals():
    etf_master_col = None


# helper: get the most recent query_date in a collection
# we use this so Parts 5 and 6 automatically work on the latest load.

def get_latest_query_date(collection):
    dates = [str(d) for d in collection.distinct("query_date") if d is not None]
    if not dates:
        raise ValueError("No query_date values found in the collection.")
    return max(dates)


# clean a stock ticker into a consistent format.
# example:
#   " aapl " -> "AAPL"
#   "BRK.B"  -> "BRKB"
# removes punctuation and spaces so holdings from different sources
# line up properly in similarity calculations.

def normalize_stock_ticker(value):
    if value is None:
        return None

    s = str(value).upper().strip()

    if s in {"", "-", "N/A", "NA", "NONE", "NULL"}:
        return None

    #then keeping letters only to standardize provider formatting differences
    s = re.sub(r"[^A-Z]", "", s)

    #rejecting the obvious bad tickers
    if not s or len(s) > 6:
        return None

    return s


#converting weight values into float.
# then handles strings like "5.23%", "5.23", or numeric inputs.
# at the end returns NaN for invalid values.

def parse_weight(value):
    if value is None:
        return np.nan

    if isinstance(value, (int, float, np.integer, np.floating)):
        return float(value)

    s = str(value).strip().replace(",", "")

    if s in {"", "-", "N/A", "NA", "NONE", "NULL"}:
        return np.nan

    s = s.replace("%", "")

    try:
        return float(s)
    except Exception:
        return np.nan

# clean one ETF's holdings list:
# - normalize stock tickers
# - make sure weights are numeric
# - merge duplicate stock tickers
# - check if the weights sum to about 100

# at last it returns:
#   cleaned_holdings, total_weight, issues
def clean_holdings_list(holdings):
    issues = []
    merged = defaultdict(float)

    if not isinstance(holdings, list):
        return [], 0.0, ["holdings_not_list"]

    for item in holdings:
        if not isinstance(item, dict):
            issues.append("holding_not_dict")
            continue

        raw_ticker = item.get("stock_ticker", item.get("ticker", item.get("symbol")))
        raw_weight = item.get("weight", item.get("weight_pct", item.get("weighting")))

        ticker = normalize_stock_ticker(raw_ticker)
        weight = parse_weight(raw_weight)

        if ticker is None:
            issues.append("bad_ticker")
            continue

        if pd.isna(weight):
            issues.append("bad_weight")
            continue

        if weight <= 0:
            issues.append("nonpositive_weight")
            continue

        merged[ticker] += float(weight)

    cleaned_holdings = [
        {"stock_ticker": ticker, "weight": round(weight, 6)}
        for ticker, weight in merged.items()
    ]
    cleaned_holdings = sorted(cleaned_holdings, key=lambda x: x["weight"], reverse=True)

    total_weight = sum(h["weight"] for h in cleaned_holdings)

    # some sources store weights as decimals summing to ~1.0 instead of ~100.0
    # if so, then scale them up to percent form.
    if 0.90 <= total_weight <= 1.10:
        for h in cleaned_holdings:
            h["weight"] = round(h["weight"] * 100.0, 6)
        total_weight = sum(h["weight"] for h in cleaned_holdings)
        issues.append("scaled_weights_from_1_to_100")

    if not cleaned_holdings:
        issues.append("no_valid_holdings")

    # project instructions want weights to sum to roughly 100%
    if not (95.0 <= total_weight <= 105.0):
        issues.append("weight_sum_outside_95_105")

    return cleaned_holdings, total_weight, sorted(set(issues))


#part 5:
#validating and optionally write cleaned holdings back into MongoDB.

def clean_and_validate_holdings_collection(etf_holdings_col, etf_master_col=None, query_date=None, write_back=True):
    if query_date is None:
        query_date = get_latest_query_date(etf_holdings_col)

    query = {"query_date": query_date}

    docs = list(
        etf_holdings_col.find(
            query,
            {
                "_id": 1,
                "etf_ticker": 1,
                "query_date": 1,
                "holdings": 1,
                "strategy_summary": 1,
            },
        )
    )

    if not docs:
        raise ValueError(f"No holdings documents found for query_date={query_date}")

    rows = []
    issue_counter = Counter()
    updates_written = 0

    for doc in docs:
        etf_ticker = str(doc.get("etf_ticker", "")).upper().strip()
        original_holdings = doc.get("holdings", [])

        cleaned_holdings, total_weight, issues = clean_holdings_list(original_holdings)

        summary = doc.get("strategy_summary")
        has_summary = isinstance(summary, str) and summary.strip() != ""

        rows.append({
            "etf_ticker": etf_ticker,
            "query_date": query_date,
            "n_holdings": len(cleaned_holdings),
            "total_weight_pct": round(total_weight, 4),
            "has_summary": has_summary,
            "issues": ", ".join(issues) if issues else "",
        })

        for issue in issues:
            issue_counter[issue] += 1

        #optionally write cleaned holdings back into MongoDB
        if write_back:
            fields_to_set = {}

            if cleaned_holdings != original_holdings:
                fields_to_set["holdings"] = cleaned_holdings

            if etf_ticker and etf_ticker != doc.get("etf_ticker"):
                fields_to_set["etf_ticker"] = etf_ticker

            if fields_to_set:
                etf_holdings_col.update_one({"_id": doc["_id"]}, {"$set": fields_to_set})
                updates_written += 1

    report_df = pd.DataFrame(rows).sort_values(
        by=["issues", "n_holdings", "etf_ticker"],
        ascending=[False, True, True]
    ).reset_index(drop=True)

    total_holdings_docs = etf_holdings_col.count_documents({"query_date": query_date})
    total_master_docs = None
    if etf_master_col is not None:
        total_master_docs = etf_master_col.count_documents({"query_date": query_date})

    print(f"Query date used: {query_date}")
    if total_master_docs is not None:
        print(f"ETF master docs on {query_date}: {total_master_docs}")
    print(f"ETF holdings docs on {query_date}: {total_holdings_docs}")
    print(f"Documents updated during cleaning: {updates_written}")
    print(f"Docs missing summaries: {(~report_df['has_summary']).sum()}")
    print(f"Docs with issue flags: {(report_df['issues'] != '').sum()}")

    print("\nIssue counts:")
    if issue_counter:
        display(
            pd.DataFrame(issue_counter.items(), columns=["issue", "count"])
            .sort_values("count", ascending=False)
            .reset_index(drop=True)
        )
    else:
        print("No issues found.")

    print("\nSample validation rows:")
    display(report_df.head(10))

    return report_df, query_date

#loading a cleaned ETF dataset from MongoDB into Python dictionaries.
#this makes similarity calculations much easier and faster.

def load_etf_dataset(etf_holdings_col, query_date):
    docs = list(
        etf_holdings_col.find(
            {"query_date": query_date, "holdings.0": {"$exists": True}},
            {"_id": 0, "etf_ticker": 1, "holdings": 1, "strategy_summary": 1, "query_date": 1}
        )
    )

    etf_data = {}

    for doc in docs:
        etf_ticker = str(doc["etf_ticker"]).upper().strip()

        holdings_map = {}
        for h in doc.get("holdings", []):
            ticker = normalize_stock_ticker(h.get("stock_ticker"))
            weight = parse_weight(h.get("weight"))

            if ticker is None or pd.isna(weight) or weight <= 0:
                continue

            holdings_map[ticker] = float(weight)

        if not holdings_map:
            continue

        etf_data[etf_ticker] = {
            "holdings_map": holdings_map,
            "holdings_set": set(holdings_map.keys()),
            "strategy_summary": (doc.get("strategy_summary") or "").strip(),
            "query_date": query_date,
        }

    if not etf_data:
        raise ValueError(f"No usable ETF data found for query_date={query_date}")

    return etf_data


#metric 1: Jaccard similarity on stock membership
# "do these two ETFs hold many of the same stocks?"

def jaccard_similarity(set1, set2):
    union = set1.union(set2)
    if not union:
        return 0.0
    return len(set1.intersection(set2)) / len(union)


#metric 2: Cosine similarity on holdings weights
# "are the weight allocations similar?"

def holdings_cosine_similarity(map1, map2):
    if not map1 or not map2:
        return 0.0

    common_keys = set(map1).intersection(map2)
    dot = sum(map1[k] * map2[k] for k in common_keys)

    norm1 = math.sqrt(sum(v * v for v in map1.values()))
    norm2 = math.sqrt(sum(v * v for v in map2.values()))

    if norm1 == 0 or norm2 == 0:
        return 0.0

    return dot / (norm1 * norm2)



#metric 3: Cosine similarity on strategy-summary embeddings
# thhis part requires Part 4 summaries to already exist in MongoDB.

def build_summary_embeddings(etf_data, model_name="nomic-ai/nomic-embed-text-v1.5"):
    summary_items = [
        (ticker, info["strategy_summary"])
        for ticker, info in etf_data.items()
        if info["strategy_summary"]
    ]

    if len(summary_items) < 2:
        print("Fewer than 2 summaries found. Summary-embedding similarity will be unavailable.")
        return {}, None

    model = SentenceTransformer(model_name, trust_remote_code=True)

    #the nomic model works better with the "search_document:" prefix
    texts = [f"search_document: {summary}" for _, summary in summary_items]

    embeddings = model.encode(
        texts,
        normalize_embeddings=True,
        show_progress_bar=True
    )

    embedding_map = {
        ticker: emb
        for (ticker, _), emb in zip(summary_items, embeddings)
    }

    return embedding_map, model


def summary_cosine_similarity(ticker1, ticker2, embedding_map):
    if ticker1 not in embedding_map or ticker2 not in embedding_map:
        return np.nan
    return float(np.dot(embedding_map[ticker1], embedding_map[ticker2]))



#helper for explanation display:
# it shows the top overlapping holdings between two ETFs

def top_shared_holdings(map1, map2, k=10):
    shared = set(map1).intersection(map2)
    ranked = sorted(shared, key=lambda t: min(map1[t], map2[t]), reverse=True)
    return ranked[:k]



# part 6:
# building recommendations for one ETF using all 3 similarity metrics.
#also returns summaries side-by-side for interpretation.

def recommend_similar_etfs(query_ticker, etf_data, embedding_map, top_n=10):
    query_ticker = str(query_ticker).upper().strip()

    if query_ticker not in etf_data:
        raise ValueError(f"{query_ticker} not found in cleaned ETF dataset.")

    query_info = etf_data[query_ticker]
    results = []

    for candidate_ticker, candidate_info in etf_data.items():
        if candidate_ticker == query_ticker:
            continue

        jac = jaccard_similarity(query_info["holdings_set"], candidate_info["holdings_set"])
        hold_cos = holdings_cosine_similarity(query_info["holdings_map"], candidate_info["holdings_map"])
        sum_cos = summary_cosine_similarity(query_ticker, candidate_ticker, embedding_map)

        # Combined score for ranking
        combined = np.nanmean([jac, hold_cos, sum_cos])

        results.append({
            "query_ticker": query_ticker,
            "candidate_ticker": candidate_ticker,
            "jaccard_similarity": round(jac, 6),
            "holdings_cosine_similarity": round(hold_cos, 6),
            "summary_cosine_similarity": round(sum_cos, 6) if not pd.isna(sum_cos) else np.nan,
            "combined_score": round(float(combined), 6) if not pd.isna(combined) else np.nan,
            "shared_holdings_count": len(query_info["holdings_set"].intersection(candidate_info["holdings_set"])),
            "top_shared_holdings": ", ".join(top_shared_holdings(query_info["holdings_map"], candidate_info["holdings_map"], k=10)),
            "query_summary": query_info["strategy_summary"],
            "candidate_summary": candidate_info["strategy_summary"],
        })

    rec_df = pd.DataFrame(results).sort_values(
        by=["combined_score", "holdings_cosine_similarity", "jaccard_similarity"],
        ascending=False
    ).reset_index(drop=True)

    return rec_df.head(top_n)

#validation against known similar ETF pairs from the assignment.

def validate_known_pairs(etf_data, embedding_map, pairs=None):
    if pairs is None:
        pairs = [("SPY", "VOO"), ("QQQ", "XLK")]

    rows = []
    for t1, t2 in pairs:
        if t1 not in etf_data or t2 not in etf_data:
            rows.append({
                "pair": f"{t1}/{t2}",
                "available": False,
                "jaccard_similarity": np.nan,
                "holdings_cosine_similarity": np.nan,
                "summary_cosine_similarity": np.nan,
            })
            continue

        rows.append({
            "pair": f"{t1}/{t2}",
            "available": True,
            "jaccard_similarity": round(jaccard_similarity(etf_data[t1]["holdings_set"], etf_data[t2]["holdings_set"]), 6),
            "holdings_cosine_similarity": round(holdings_cosine_similarity(etf_data[t1]["holdings_map"], etf_data[t2]["holdings_map"]), 6),
            "summary_cosine_similarity": round(summary_cosine_similarity(t1, t2, embedding_map), 6)
            if not pd.isna(summary_cosine_similarity(t1, t2, embedding_map)) else np.nan,
        })

    return pd.DataFrame(rows)

# it prints top names by each metric so you can discuss
# where the metrics agree and where they disagree.

def compare_metric_rankings(rec_df, top_n=5):
    metric_cols = [
        "jaccard_similarity",
        "holdings_cosine_similarity",
        "summary_cosine_similarity",
        "combined_score",
    ]

    for metric in metric_cols:
        top_names = rec_df.sort_values(metric, ascending=False)["candidate_ticker"].head(top_n).tolist()
        print(f"Top {top_n} by {metric}: {top_names}")


# RUN PART 5

quality_report, QUERY_DATE = clean_and_validate_holdings_collection(
    etf_holdings_col=etf_holdings_col,
    etf_master_col=etf_master_col,
    query_date=None,     # use latest available date
    write_back=True      # set False if you only want to audit without updating Mongo
)


# RUN Ppart 6

etf_data = load_etf_dataset(etf_holdings_col, QUERY_DATE)
embedding_map, embedding_model = build_summary_embeddings(etf_data)

print(f"\nUsable ETFs for similarity on {QUERY_DATE}: {len(etf_data)}")
print(f"ETFs with strategy-summary embeddings: {len(embedding_map)}")

print("\nKnown-pair validation:")
pair_validation = validate_known_pairs(etf_data, embedding_map)
display(pair_validation)

#testing examples requested in the assignment
test_queries = ["SPY", "QQQ", "ARKK"]

for q in test_queries:
    if q not in etf_data:
        print(f"\n{q} not found for query_date={QUERY_DATE}; skipping.")
        continue

    print(f"\nTop recommendations for {q}")
    recs = recommend_similar_etfs(q, etf_data, embedding_map, top_n=10)

    display(
        recs[
            [
                "candidate_ticker",
                "jaccard_similarity",
                "holdings_cosine_similarity",
                "summary_cosine_similarity",
                "combined_score",
                "shared_holdings_count",
                "top_shared_holdings",
                "candidate_summary",
            ]
        ]
    )

    print(f"Metric agreement/disagreement for {q}:")
    compare_metric_rankings(recs, top_n=5)
````
